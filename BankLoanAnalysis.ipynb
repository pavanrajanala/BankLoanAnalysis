```python
df['Loan_Status'].value_counts().plot(kind='bar')
plt.title('Loan Status')
plt.xlabel('Staeus')
plt.ylabel('Count')
plt.show()
```

```python
avg_income = np.mean(df['ApplicantIncome'])
print("avg_income : ",avg_income)
median_income = np.median(df['ApplicantIncome'])
print("median_income : ",median_income)
std_income = np.std(df['ApplicantIncome'])
print("std_income : ",std_income)
min_income = np.min(df['ApplicantIncome'])
print("min_income : ",min_income)
max_income = np.max(df['ApplicantIncome'])
print("max_income : ",max_income)
min_loan = np.min(df['LoanAmount'])
print("Min Loan : ",min_loan)
max_loan=np.max(df['LoanAmount'])
print("Max Laon : ",max_loan)
std_loan=np.std(df['LoanAmount'])
print("Std Loan : ",std_loan)
median_loan=np.median(df['LoanAmount'])
print("Median Loan : ",median_loan)
```

```python
p25 = np.percentile(df['ApplicantIncome'],25)
p50 = np.percentile(df['ApplicantIncome'],50)
p75 = np.percentile(df['ApplicantIncome'],75)

print(p25,p50,p75)
```
